```python
import os

USER_FILE = 'users.txt'

def load_users():
    users = {}
    if os.path.exists(USER_FILE):
        with open(USER_FILE, 'r') as f:
            for line in f:
                name, email, address, password = line.strip().split(', ')
                users[email] = {'name': name, 'email': email, 'address': address, 'password': password}
    return users

def save_user(user):
    with open(USER_FILE, 'a') as f:
        f.write(f"{user['name']}, {user['email']}, {user['address']}, {user['password']}\n")

def register(users):
    print("\n=== Register ===")
    name = input("Name: ")
    email = input("Email: ")
    if email in users:
        print("Email already registered.")
        return
    address = input("Address: ")
    password = input("Password: ")
    user = {'name': name, 'email': email, 'address': address, 'password': password}
    users[email] = user
    save_user(user)
    print("Registration successful!")

def login(users):
    print("\n=== Login ===")
    email = input("Email: ")
    password = input("Password: ")
    user = users.get(email)
    if user and user['password'] == password:
        print(f"\nWelcome, {user['name']}!")
        print("Your Profile:")
        print(f"Name: {user['name']}")
        print(f"Email: {user['email']}")
        print(f"Address: {user['address']}")
    else:
        print("Invalid email or password.")

def main():
    users = load_users()
    while True:
        print("\n1. Register")
        print("2. Login")
        print("3. Exit")
        choice = input("Choose an option: ")
        if choice == '1':
            register(users)
        elif choice == '2':
            login(users)
        elif choice == '3':
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
